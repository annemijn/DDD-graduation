I added new data based on the research on new data of customers values and goals in the academic paper.
```python
def review_score (row): #create function
   if row['provider name'] == 0: #if provider name = BNP paribas, return a review score of 76.
      return 76
   if row['provider name'] == 1:
      return 79
   if row['provider name'] == 2:
      return 87
   if row['provider name'] == 3:
      return 77
   if row['provider name'] == 4:
      return 74
   if row['provider name'] == 5:
      return 90   
   if row['provider name'] == 6:
      return 78 
   if row['provider name'] == 7:
      return 79        
   return 0
```

### Values and goals
Then the variables consumption values and goals are assigned to the customer based on their data. 

```python
def consumption_value (row): #create function
   if row['preference compare products'] == True: #if customer compares products, return value 1.
      return 1
   if row['position product'] == 1 or row['sorted by'] == 2:
      return 2
   if row['preference compare products'] == True and row['sorted by'] == 2 or row['position product'] == 1:
      return 3
   return 0
```

```python
df_subset.apply (lambda row: consumption_value(row), axis=1) #apply the function to the df
```

```python
df_subset['consumption values'] = df_subset.apply (lambda row: consumption_value(row), axis=1) #add new variable consumption values with the values of the outcome of the function.
df_subset.head()
```

```python
def goals (row):#create function
   if row['loan duration'] <= 24: #if customer ha a loan with a loan duration of 24 or lower, return value 1.
      return 1        
   if row['preference compare products'] == 1:
      return 2
   if row['review score'] >= 70:
      return 3      
   return 0
```

```python
df_subset.apply (lambda row: goals(row), axis=1) #apply the function to the df
```

```python
df_subset['goals'] = df_subset.apply (lambda row: goals(row), axis=1)#add new variable goals with the values of the outcome of the function.
df_subset.head(5000)
```
