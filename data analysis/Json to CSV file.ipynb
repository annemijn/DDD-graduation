For this study, a dataset is provided from the finance company Geld.nl. The dataset contains behavioral data of customers that have applied for a loan on the Geld.nl website.The dataset that has been supplied contains data from March 2019 till May 2020. The datasets are delivered in separate files per week and are in JSON format. In this Notebook, the different data files are merged into one dataset and converted into a CSV format.

```python
import pandas as pd
import numpy as np 
import json 
from pandas.io.json import json_normalize #package for flattening json in pandas df
```

----
## Normalizing Json files

First I loaded the json files into the Notebook. Then I created different dataframes with the normalized json files.

```python
#load json object
with open('data4/1.json',encoding='utf-8') as f:
    d = json.load(f)

#lets put the data into a pandas df
df = json_normalize(d) #normalize json data into dataframe
df.head() #show the resulting data
```

```python
#load json object
with open('data4/2.json',encoding='utf-8') as f:
    d2 = json.load(f)

#lets put the data into a pandas df
df2 = json_normalize(d2) #normalize json data into dataframe
df2.head() #show the resulting data
```

```python
#load json object
with open('data4/3.json',encoding='utf-8') as f:
    d3 = json.load(f)

#lets put the data into a pandas df
df3 = json_normalize(d3) #normalize json data into dataframe
df3.head() #show the resulting data
```

```python
#load json object
with open('data4/4.json',encoding='utf-8') as f:
    d4 = json.load(f)

#lets put the data into a pandas df
df4 = json_normalize(d4) #normalize json data into dataframe
df4.head() #show the resulting data
```

Then I put the different dataframes into one dataframe.

```python
pdList = [df, df2, df3, df4]
df_subset = pd.concat(pdList)

df_subset.head()
```
